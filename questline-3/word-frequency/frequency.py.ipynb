```python
#frequency.py

f = open("sample_input.txt", "r")
txt = f.read()
f.close()

w= txt.lower().split()

# Count words using a dictionary
wcount = {}
for i in w:
    if i in wcount:
        wcount[i] += 1
    else:
        wcount[i] = 1

f1 = open("output.txt", "w")
for w, count in wcount.items():
    l = f"{w} -> {count}"
    print(l)
    f1.write(l + "\n")

f1.close()
```
